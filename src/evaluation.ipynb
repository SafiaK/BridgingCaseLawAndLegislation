```python
from sklearn.metrics import precision_score, recall_score, f1_score
df_sections['contains_application'] = df_sections['contains_application'].map({'yes': 1, 'no': 0})
#df_sections['if_law_applied_actual'] = df_sections['if_law_applied_actual'].map({'yes': 1, 'no': 0})

y_pred = df_sections['contains_application']
y_true = df_sections['if_law_applied_actual']

precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print(f"Precision: {precision}, Recall: {recall}, F1-score: {f1}")
```
