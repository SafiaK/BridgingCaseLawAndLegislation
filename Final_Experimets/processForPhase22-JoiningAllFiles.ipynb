````python
# Combine three JSONL files into one
input_file1 = "../data/final_test/extraction_results/batch_law_application_output1_with_gpt_4o.jsonl"
input_file2 = "../data/final_test/extraction_results/batch_law_application_output2_with_gpt_4o.jsonl"

output_file = "../data/final_test/extraction_results/batch_law_application_output_with_gpt_4o.jsonl"

# Read and combine the files
with open(output_file, 'w', encoding='utf-8') as outfile:
    # Process first file
    with open(input_file1, 'r', encoding='utf-8') as infile1:
        for line in infile1:
            outfile.write(line)
    
    # Process second file
    with open(input_file2, 'r', encoding='utf-8') as infile2:
        for line in infile2:
            outfile.write(line)
            


print(f"Combined file saved to: {output_file}")

# Verify the number of lines in each file
with open(input_file1, 'r', encoding='utf-8') as f1:
    lines1 = sum(1 for _ in f1)
with open(input_file2, 'r', encoding='utf-8') as f2:
    lines2 = sum(1 for _ in f2)

with open(output_file, 'r', encoding='utf-8') as f4:
    lines4 = sum(1 for _ in f4)

print(f"Number of lines in first file: {lines1}")
print(f"Number of lines in second file: {lines2}")
print(f"Total lines in combined file: {lines4}")
````

````python
#join this sampleoutput with the original input give
import pandas as pd
import json

# The original DataFrame 'df' (with 20 rows) should be in memory from cell 6.
# If not, uncomment the next line to reload it:
# df = pd.read_csv('../data/final_test/combined_all_csvfiles.csv', nrows=20)

# Path to the batch output file you downloaded
output_file_path = "../data/final_test/extraction_results/batch_law_application_output1_with_gpt_4o.jsonl"

# List to store parsed results from the batch output
parsed_results = []

with open(output_file_path, 'r') as f:
    for line in f:
        data = json.loads(line)
        
        # The custom_id (e.g., 'request_1') links back to the original row index
        custom_id = data.get('custom_id')
        # Extract the number from 'request_x' and convert to a 0-based index
        original_index = int(custom_id.split('_')[-1]) - 1
        
        try:
            # The model's response is a JSON string inside the 'content' field
            content_str = data['response']['body']['choices'][0]['message']['content']
            content_str = content_str.replace("```json", "").replace("```", "")
            extracted_data = json.loads(content_str)
            
            # Store the extracted data with its original index
            parsed_results.append({
                'original_index': original_index,
                'extracted_phrases': extracted_data
            })
            
        except (KeyError, IndexError, json.JSONDecodeError) as e:
            print(f"Could not parse result for {custom_id}: {e}")
            parsed_results.append({
                'original_index': original_index,
                'extracted_phrases': [{'error': str(e)}]
            })

# Create a DataFrame from the parsed batch results
output_df = pd.DataFrame(parsed_results)

# Join the results back to the original input DataFrame using the index
merged_df = df.merge(output_df, left_index=True, right_on='original_index')

# Display the result
print("Successfully joined batch results with the original data.")
merged_df.head()
````

````python
# Combine rows so that each paragraph ('para_id') has a single row
# with a dictionary containing all its associated sections.

# 1. Create a new DataFrame by grouping by 'para_id'.
#    For each group, create a dictionary of {section_id: section_text}.
sections_dict_df = merged_df.groupby('para_id').apply(
    lambda g: dict(zip(g['section_id'], g['section_text']))
).reset_index(name='sections')

# 2. Get the rest of the data for each paragraph.
#    We drop the section columns and then remove duplicate rows for each 'para_id',
#    keeping only the *first* appearance of each paragraph.
other_cols_df = merged_df.drop(columns=['section_id', 'section_text']).drop_duplicates(subset=['para_id'])

# 3. Merge the two DataFrames back together on 'para_id'.
combined_df = pd.merge(other_cols_df, sections_dict_df, on='para_id')

# Display the head of the newly structured DataFrame
print("DataFrame after combining rows per paragraph:")
combined_df.head()
````
